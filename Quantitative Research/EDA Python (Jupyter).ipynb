```python
#-----------------------------------------------EXPLORATORY DATA ANALYSIS BY BADAR WIRYAWAN-------------------------------------------------

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#-----------------------------------------------EXPLORATORY DATA ANALYSIS BY BADAR WIRYAWAN-------------------------------------------------

file_path = input("Masukkan Lokasi File (Klik kanan > Copy as Path): ")
independent_var = input("Masukkan Variabel Independen (var1, var2, var3): ")
dependent_var = input("Masukkan Variabel Dependen: ")
Visualization = input("Pilih visualisasi (1: Histogram, 2: Boxplot, 3: Jointplot, 4: Lmplot, 5: Pearson Correlation Matrix): ")

#-------------CONTOH INPUT-------------
#file_path = D:\Programming\\advertising.csv Sumber: https://www.kaggle.com/datasets/mehmetisik/advertisingcsv/data
#independent_var = TV, radio, newspaper
#dependent_var = sales
#Visualization = 1

#Dengan input di atas, program akan memunculkan profil data dalam bentuk Histogram (1).

#-----------------------------------------------EXPLORATORY DATA ANALYSIS BY BADAR WIRYAWAN-------------------------------------------------

#data = pd.read_csv(file_path) # ---> Untuk file CSV.
data = pd.read_excel(file_path) # ---> Untuk file Excel (.xlsx).

independent_list = independent_var.split(", ")
var_list = independent_list.append(dependent_var)
independent_list00 = independent_var.split(", ")
dependent_list = []
dependent_list.append(dependent_var)

#-----------------------------------------------EXPLORATORY DATA ANALYSIS BY BADAR WIRYAWAN-------------------------------------------------

def histogram(file=data):
    for a in independent_list:
        sns.histplot(file[a], kde=True)
        plt.show()
    print(data[independent_list].describe().T)

def boxtplot(file=data):
    for b in independent_list:
        sns.boxplot(file[b])
        plt.show()
    print(data[independent_list].describe().T)

def jointplot(file=data):
    for c in independent_list00:
        sns.jointplot(data=file, x=c, y=dependent_var, kind="reg")
        plt.show()

def lmplot(file=data):
    for d in independent_list00:
        sns.lmplot(data=file, x=d, y=dependent_var)
        plt.show()

def heatmap(file=data):
    new_data = data[independent_list]
    sns.heatmap(new_data.corr(method="pearson"), cmap="gray", annot=True).set_title(label="Pearson Correlation Matrix")
    plt.show()

#-----------------------------------------------EXPLORATORY DATA ANALYSIS BY BADAR WIRYAWAN-------------------------------------------------

if Visualization == "1":
    print(histogram())
elif Visualization == "2":
    print(boxtplot())
elif Visualization == "3":
    print(jointplot())
elif Visualization == "4":
    print(lmplot())
elif Visualization == "5":
    print(heatmap())

#-----------------------------------------------EXPLORATORY DATA ANALYSIS BY BADAR WIRYAWAN-------------------------------------------------
```
